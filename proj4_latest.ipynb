```python
def get_summary(new_ref):
    global jobtype
    print(new_ref)
    new_page = requests.get(new_ref)
    newsoup = BeautifulSoup(new_page.content,"lxml")
    
    try:
        summary = newsoup.find('div',{"class":"templatetext"}).text
        jobtype = newsoup.find('dd',{"data-automation":"job-detail-work-type"}).text
        
        
    except:
        summary = newsoup.find('div',{"data-automation":"mobileTemplate"}).text
        jobtype = newsoup.find('dd',{"data-automation":"job-detail-work-type"}).text
       
    return summary
    
```
